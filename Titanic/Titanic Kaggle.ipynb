```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
submission = pd.read_csv('gender_submission.csv')
```

```python
train_data.head()
```

```python
train_data.info()
```

* **Age**, **Cabin** and **Embarked** has incomplete data, MISSING VALUE TREATMENT required.

```python
train_data['Embarked'].unique()
```

```python
train_data.groupby('Embarked')['PassengerId'].nunique()
```

* Max of the Passengers Embarked from 'S'
* Filling up the Missing values in Embarked column, with 'S'.

```python
train_data['Embarked'].fillna('S', inplace = True)
```

```python
train_data['Cabin'].fillna('X', inplace = True)
train_data['CabinX'] = train_data['Cabin'].apply(lambda x: str(x)[0])
```

```python
train_data.groupby('CabinX')['Fare'].mean()
```

* Grouping the Cabin Data by Fare.
* Converting the Cabin Data to Ordinal Scale.

```python
train_data['CabinX'].replace(['B', 'C'], 3, inplace=True)
train_data['CabinX'].replace(['A', 'D', 'E', 'T'], 2, inplace=True)
train_data['CabinX'].replace(['F', 'G'], 1, inplace=True)
train_data['CabinX'].replace(['X'], 0, inplace=True)
```

```python
ax = plt.axes()

data = train_data.groupby(['CabinX'])[['Survived']].count()
sns.barplot(x = data.index,
            y = data['Survived'],
            alpha = 0.8,
            color = 'violet',
            ax = ax)

data = train_data.groupby(['CabinX'])[['Survived']].sum()
sns.barplot(x=data.index, 
            y=data['Survived'], 
            alpha=0.8, 
            color='cornflowerblue', 
            ax=ax)

ax.set_title('Survived or Not')
sns.plt.show()
```

```python
train_data['SexX'] = train_data['Sex'].map({'female': 1, 'male':0})
train_data['PclassX'] = train_data['Pclass'].map({1: 3, 2: 2, 3:1})
```

```python
train_data.loc[train_data['SexX']==0, 'SexByPclass'] = train_data.loc[train_data['SexX']==0, 'PclassX']
train_data.loc[train_data['SexX']==1, 'SexByPclass'] = train_data.loc[train_data['SexX']==1, 'PclassX'] + 3
train_data['SexByPclass'] = train_data['SexByPclass'].astype(int)
```
